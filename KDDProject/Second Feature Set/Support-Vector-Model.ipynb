```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv('round.csv')
```

```python
data['c2_familiarity_r1'].fillna((data['c2_familiarity_r1'].mean()), inplace=True)
data['c2_familiarity_r2'].fillna((data['c2_familiarity_r2'].mean()), inplace=True)
data['c2_familiarity_r3'].fillna((data['c2_familiarity_r3'].mean()), inplace=True)
```

```python
data = data[['second_rating','like_avg','c1_familiarity_r1','c1_familiarity_r2','c1_familiarity_r3','c2_familiarity_r1','c2_familiarity_r2','c2_familiarity_r3','surprise_avg']] 
data
```

```python
data["target"]= data["surprise_avg"].apply(lambda x: 0 if x < 3 else 1  )
data
```

```python
X_train, X_test, y_train, y_test = train_test_split(data.drop(['surprise_avg','target'],axis=1), 
                                                    data['target'], test_size=0.30, 
                                                    random_state=101)
```

```python
from sklearn.svm import SVC
model = SVC()
model.fit(X_train,y_train)
predictions = model.predict(X_test)
from sklearn.metrics import classification_report
print(classification_report(y_test,predictions))
```
