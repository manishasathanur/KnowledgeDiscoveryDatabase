```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv('second.csv')
data
```

```python
data = data[['frequency','like','c1_familiarity','surprise']] 
data
```

```python
data["target"]= data["surprise"].apply(lambda x: 0 if x < 3 else 1  )
data
```

```python
X_train, X_test, y_train, y_test = train_test_split(data.drop(['surprise','target'],axis=1), 
                                                    data['target'], test_size=0.30, 
                                                    random_state=101)
```

```python
from sklearn.svm import SVC
model = SVC()
model.fit(X_train,y_train)
predictions = model.predict(X_test)
from sklearn.metrics import classification_report
print(classification_report(y_test,predictions))
```
