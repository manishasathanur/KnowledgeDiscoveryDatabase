```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import time
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv("round.csv")
```

```python
data
```

```python
data = data[['second_rating','like_avg','c1_familiarity_r1','c1_familiarity_r2','c1_familiarity_r3','surprise_avg']] 
data
```

```python
data["target"]= data["surprise_avg"].apply(lambda x: 0 if x < 3 else 1  )
```

```python
data
```

```python
X_train, X_test, y_train, y_test = train_test_split(data.drop(['surprise_avg','target'],axis=1), 
                                                    data['target'], test_size=0.30, 
                                                    random_state=101)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train,y_train)
y_pred = knn.predict(X_test)
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
